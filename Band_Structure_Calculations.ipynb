```python
from scipy.linalg import eigvals
def calculate_exp_sum(k, d):
    return np.exp(np.sum(1j * k * d))

def H(P1, P2, Nt):

  a = 3.18
  a1 = a * np.array([1 , 0])
  a2 = a * np.array([-1/2 , np.sqrt(3) / 2])
  d1 = a1
  d2 = a1 + a2
  d3 = a2
  d4 = - ((2 * a1) + a2) / 3
  d5 = (a1 + (2*a2)) / 3
  d6 = (a1 - a2) / 3
  d7 = -2 * (a1 + (2 * a2)) / 3
  d8 = 2 * ((2*a1)+a2) / 3
  d9 = 2 * (a2 - 1) / 3;

  kx = np.linspace(P1[0], P2[0], Nt)
  ky = np.linspace(P1[1], P2[1], Nt)

  Eig = []

  # Calculating the Hamiltonian matrix
  for Nk in range(Nt):
      k = np.array([kx[Nk], ky[Nk]])
      H1 = np.zeros((11, 11), dtype = complex)
      H2 = np.zeros((11, 11), dtype = complex)
      H2[8, 5] = t6[8, 5] * (calculate_exp_sum(k, d7) + calculate_exp_sum(k, d8) + calculate_exp_sum(k, d9))
      H2[10, 5] = t6[10, 5] * (calculate_exp_sum(k, d7) - calculate_exp_sum(k, d8) / 2 - calculate_exp_sum(k, d9) / 2)
      H2[9, 5] = np.sqrt(3) * t6[10, 5] * (-calculate_exp_sum(k, d8) + calculate_exp_sum(k, d9)) / 2
      H2[8, 7] = t6[8, 7] * (calculate_exp_sum(k, d7) - calculate_exp_sum(k, d8) / 2 - calculate_exp_sum(k, d9) / 2)
      H2[7, 8] = np.sqrt(3) * t6[8, 7] * (-calculate_exp_sum(k, d8) + calculate_exp_sum(k, d9)) / 2
      H2[9, 7] = 3 * t6[10, 7] * (calculate_exp_sum(k, d8) + calculate_exp_sum(k, d9)) / 4
      H2[10, 7] = np.sqrt(3) * t6[10, 7] * (calculate_exp_sum(k, d8) - calculate_exp_sum(k, d9)) / 4
      H2[9, 8] = H2[8, 7]
      H2[10, 8] = H2[10, 7]
      H2[8, 10] = H2[10, 7]
      H2[10, 10] = t6[10, 10] * (calculate_exp_sum(k, d7) + calculate_exp_sum(k, d8) / 4 + calculate_exp_sum(k, d9) / 4)
      H2[5, 8] = H2[8, 5]
      H2[5, 10] = H2[10, 5]
      H2[5, 9] = H2[9, 5]
      H2[7, 8] = H2[8, 7]
      H2[7, 9] = H2[9, 7]
      H2[7, 10] = H2[10, 7]
      H2[7, 5] = H2[5, 7]
      H2[8, 10] = H2[10, 8]
      H2[8, 5] = H2[5, 8]
      H2[9, 10] = H2[10, 9]
      H2[9, 5] = H2[5, 9]
      H2[10, 9] = H2[9, 10]
      for i in range(11):
          for j in range(11):
              if i == j:
                  H1[i, j] = E[i] + (2 * t1[i, i] * np.cos(np.sum(k * d1))) + (2 * t2[i, i] * (np.cos(np.sum(k * d2)) + np.cos(np.sum(k * d3))))
              if ((i == 2 and j == 4) or (i == 5 and j == 7) or (i == 8 and j == 10)):
                  H1[i, j] = 2 * t1[i, j] * np.cos(np.sum(k * d1)) + \
                            t2[i, j] * (np.exp(np.sum(-1j * k * d2)) + np.exp(np.sum(-1j * k * d3))) + \
                            t3[i, j] * (np.exp(np.sum(1j * k * d2)) + np.exp(np.sum(1j * k * d3)))
                  H1[j, i] = np.conj(H1[i, j])

              if ((i == 0 and j == 1) or (i == 2 and j == 3) or (i == 3 and j == 4) or
                  (i == 5 and j == 6) or (i == 6 and j == 7) or (i == 8 and j == 9) or (i == 9 and j == 10)):
                  H1[i, j] = -2j * t1[i, j] * np.sin(np.sum(k * d1)) + \
                            t2[i, j] * (np.exp(np.sum(-1j * k * d2)) - np.exp(np.sum(-1j * k * d3))) + \
                            t3[i, j] * (-np.exp(np.sum(1j * k * d2)) + np.exp(np.sum(1j * k * d3)))
                  H1[j, i] = np.conj(H1[i, j])

              if ((i == 2 and j == 0) or (i == 4 and j == 0) or (i == 3 and j == 1) or
                  (i == 9 and j == 5) or (i == 8 and j == 6) or (i == 10 and j == 6) or (i == 9 and j == 7)):
                  H1[i, j] = t4[i, j] * (np.exp(np.sum(1j * k * d4)) - np.exp(np.sum(1j * k * d6)))
                  H1[j, i] = np.conj(H1[i, j])

              if ((i == 3 and j == 0) or (i == 2 and j == 1) or (i == 4 and j == 1) or
                  (i == 8 and j == 5) or (i == 10 and j == 5) or (i == 9 and j == 6) or
                  (i == 8 and j == 7) or (i == 10 and j == 7)):
                  H1[i, j] = t4[i, j] * (np.exp(np.sum(1j * k * d4)) + np.exp(np.sum(1j * k * d6))) + \
                            t5[i, j] * np.exp(np.sum(1j * k * d5))
                  H1[j, i] = np.conj(H1[i, j])

      H_total = H1 + H2
      eigenvalues = eigvals(H_total)
      Eig.append(np.sort(np.real(eigenvalues)))
  return Eig
```

```python
a = 3.18
G = [0, 0]
M = [np.pi / a, np.pi/np.sqrt(3)/a]
K = [4 * np.pi / 3 / a, 0]
distance_G_to_M = np.sqrt((G[0] - M[0])**2 + (G[1] - M[1])**2)
distance_M_to_K = np.sqrt((M[0] - K[0])**2 + (M[1] - K[1])**2)
distance_K_to_G = np.sqrt((K[0] - G[0])**2 + (K[1] - G[1])**2)

N1 = int(100 * distance_G_to_M)
N2 = int(100 * distance_M_to_K)
N3 = int(100 * distance_K_to_G)
eigenvalues = []
eigenvalues.extend(H(G, M, N1))
eigenvalues.extend(H(M, K, N2))
eigenvalues.extend(H(K, G, N3))
```

```python
import matplotlib.pyplot as plt

N = 100
tick_label_font_size = 20
axis_label_font_size = 20
eigenvalues = np.array(eigenvalues)
path = np.arange(N1 + N2 + N3)
# Plot the eigenvalues
plt.figure(figsize=(10, 6))
for i in range(eigenvalues.shape[1]):
    plt.plot(path, eigenvalues[:, i], label=f'Eigenvalue {i+1}', linewidth = 2)
plt.xticks([path[0], path[N1-1], path[N1 + N2 -1], path[N1 + N2 + N3-1]],
           [r'$\Gamma$', 'M', 'K', '$\Gamma$'], fontsize=tick_label_font_size)
plt.yticks(fontsize=tick_label_font_size)
plt.xlabel('Path', fontsize=axis_label_font_size)
plt.ylabel('E (eV)', fontsize=axis_label_font_size)
#plt.legend()
plt.ylim([-2,4])
plt.title("Band Structure of Monolayer MoS2 from Tightbinding Model",fontsize= 17)
plt.show()
```

```python
d =  0.05
K_sub = [4 * np.pi / 3 / a - d,  0]
K_plus = [4 * np.pi / 3 / a + d, 0]

NN = 100
K_x = np.linspace(4 * np.pi / 3 / a - d, 4 * np.pi / 3 / a + d, NN)
K_eigenvalues = H(K_sub, K_plus, NN)
K_eigenvalues = np.array(K_eigenvalues)
for i in range(K_eigenvalues.shape[1]):
    plt.plot(K_x, K_eigenvalues[:, i], label=f'Eigenvalue {i+1}', linewidth = 2)

plt.legend()
plt.ylim([-2,4])
plt.show()
```

```python
from scipy.optimize import curve_fit
def quadratic(x, a, b, c):
    return a * x**2 + b * x + c
index = 6
# Perform the quadratic fit
params, params_covariance = curve_fit(quadratic, K_x, K_eigenvalues[:, index])
plt.figure(figsize=(8, 6))
plt.scatter(K_x, K_eigenvalues[:, index], label='Data')
plt.plot(K_x, quadratic(K_x, *params), label='Quadratic fit: a=%5.3f, b=%5.3f, c=%5.3f' % tuple(params))
plt.xlabel('K_x')
plt.ylabel('y')
plt.legend()
plt.show()
```
